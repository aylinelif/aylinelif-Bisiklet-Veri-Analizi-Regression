# Veriyi test ve train olmak Ã¼zere ikiye ayÄ±rmak

```python
from sklearn.model_selection import train_test_split
```

```python
## train_test_split
```

```python
y = dataFrame["Fiyat"].values #bir numpy dizisine Ã§evirdik
```

#y = wx + b
y -> label


```python
x = dataFrame[["BisikletOzellik1","BisikletOzellik2"]].values
```

x -> featur(Ã¶zellik)

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=15)
```

 # Scaling (boyutunu bÃ¼yÃ¼tmek veya kÃ¼Ã§Ã¼ltmek)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
import tensorflow as tf 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add (Dense(5,activation="relu"))
model.add (Dense(5,activation="relu"))
model.add (Dense(5,activation="relu"))

model.add (Dense(1)) #bir Ã§Ä±ktÄ± nÃ¶ronu olmasÄ± iÃ§in bazÄ± karmaÄ±ÅŸÄ±k problemlerde birden fazla Ã§Ä±ktÄ± nÃ¶ronu olabilir

model.compile(optimizer = "rmsprop", loss = "mse")
```

```python
model.fit(x_train,y_train,epochs=250)
```

```python
loss = model.history.history["loss"]
```

Bu kod, modelin eÄŸitim sÃ¼reci boyunca kaydedilen kayÄ±p (loss) deÄŸerlerini alÄ±r ve loss isimli bir deÄŸiÅŸkene atar.

DetaylÄ± AÃ§Ä±klama

model.history:

model.fit() fonksiyonu Ã§alÄ±ÅŸtÄ±ÄŸÄ±nda geriye bir History objesi dÃ¶ner.

model.history.history:

Bu History nesnesinin iÃ§inde eÄŸitim sÃ¼recinde elde edilen deÄŸerler (loss, accuracy vb.) sÃ¶zlÃ¼k (dictionary) ÅŸeklinde tutulur.

"loss":

Bu sÃ¶zlÃ¼k iÃ§inden her epoch sonunda kaydedilen loss deÄŸerlerini seÃ§er.

```python
sbn.lineplot(x=range(len(loss)),y=loss)
```

Bu komut, Seaborn kÃ¼tÃ¼phanesi kullanÄ±larak loss deÄŸerlerinin zaman iÃ§indeki deÄŸiÅŸimini Ã§izgi grafik (line plot) olarak gÃ¶rselleÅŸtirir.loss	Her epoch iÃ§in kaydedilen kayÄ±p deÄŸerleri listesidir.

y=loss	Y eksenine loss deÄŸerleri Ã§izilir.
