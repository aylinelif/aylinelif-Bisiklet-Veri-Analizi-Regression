# Veriyi test ve train olmak üzere ikiye ayırmak

```python
from sklearn.model_selection import train_test_split
```

```python
## train_test_split
```

```python
y = dataFrame["Fiyat"].values #bir numpy dizisine çevirdik
```

#y = wx + b
y -> label


```python
x = dataFrame[["BisikletOzellik1","BisikletOzellik2"]].values
```

x -> featur(özellik)

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=15)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_test
```

```python
y_test.shape
```

```python
y_train.shape
```

 # Scaling (boyutunu büyütmek veya küçültmek)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
import tensorflow as tf 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add (Dense(5,activation="relu"))
model.add (Dense(5,activation="relu"))
model.add (Dense(5,activation="relu"))

model.add (Dense(1)) #bir çıktı nöronu olması için bazı karmaışık problemlerde birden fazla çıktı nöronu olabilir

model.compile(optimizer = "rmsprop", loss = "mse")
```

```python
model.fit(x_train,y_train,epochs=250)
```

```python
loss = model.history.history["loss"]
```

Bu kod, modelin eğitim süreci boyunca kaydedilen kayıp (loss) değerlerini alır ve loss isimli bir değişkene atar.

Detaylı Açıklama

model.history:

model.fit() fonksiyonu çalıştığında geriye bir History objesi döner.

model.history.history:

Bu History nesnesinin içinde eğitim sürecinde elde edilen değerler (loss, accuracy vb.) sözlük (dictionary) şeklinde tutulur.

"loss":

Bu sözlük içinden her epoch sonunda kaydedilen loss değerlerini seçer.

```python
sbn.lineplot(x=range(len(loss)),y=loss)
```

Bu komut, Seaborn kütüphanesi kullanılarak loss değerlerinin zaman içindeki değişimini çizgi grafik (line plot) olarak görselleştirir.loss	Her epoch için kaydedilen kayıp değerleri listesidir.
range(len(loss))	X eksenine epoch numaralarını verir. (0,1,2,3,... şeklinde)
y=loss	Y eksenine loss değerleri çizilir.
sbn.lineplot()	Seaborn ile çizgi grafik oluşturur.

```python
trainLoss = model.evaluate (x_train,y_train,verbose = 0) #tanımlanmış bir Keras/TensorFlow modeli için eğitim verileri üzerinde modelin performansını değerlendirmek amacıyla kullanılır
```

```python
testLoss = model.evaluate (x_test,y_test,verbose = 0)
```

```python
trainLoss 
```

```python
testLoss #trainLoss ve testLoss birbirine ne kadar yakınsa o kadar iyidir
```

model.evaluate() fonksiyonu:

Modele verilen veri üzerinde kayıp (loss) ve tanımlanmış metrikleri hesaplar.

Eğitimde kullanılan veriler üzerinde modelin başarısını görmek için kullanılır.

x_train, y_train:

Modelin değerlendirilmesini istediğiniz giriş ve çıkış verileridir.

Bu kodda eğitim verileriyle değerlendirildiği için trainLoss modelin eğitim setine ne kadar iyi uyduğunu gösterir.

verbose=0:

Ekrana herhangi bir çıktı bastırılmasını engeller.
